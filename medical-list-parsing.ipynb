### Use this module do dump CSV medicine list to brandnames JSON file ###

```python
import pandas as pd
import json
```

```python
cs = pd.read_csv("med-list-extended.csv", dtype=str)
cs = cs[["code", "name", "price"]]
cs["price"] = cs["price"].astype(float)
```

```python
vallist = cs["name"].values.tolist()
vallist
```

```python
with open("pharping-medicine-li.json", 'w') as pdr:
   pdr.write( json.dumps(vallist))
#     pdr.write(str(cs["name"].values.tolist()))
pdr.close()
```
